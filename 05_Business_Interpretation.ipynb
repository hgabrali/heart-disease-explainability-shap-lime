**Best Model:** XGBoost (ROC-AUC ~ 0.92)  
**Explainability Tools:** SHAP (global + local) and LIME (local)

## Top 5 Predictors and Their Business Meaning

| Rank | Feature | What It Measures | Business Insight |
|------|---------|-----------------|-----------------|
| 1 | cp — Chest Pain Type | Type of chest pain (typical, atypical, non-anginal, asymptomatic) | Asymptomatic chest pain is paradoxically the **highest-risk** pattern — silent ischemia is dangerous |
| 2 | thalach — Max Heart Rate | Peak heart rate achieved during exercise stress test | **Lower max HR** during exercise is a red flag — the heart cannot respond to demand |
| 3 | ca — Major Vessels | Number of major coronary vessels coloured by fluoroscopy (0-3) | Every additional blocked vessel significantly raises disease probability |
| 4 | oldpeak — ST Depression | ST segment depression during exercise vs rest (ECG reading) | Measures ischemia severity — **higher values = more oxygen debt** in heart muscle |
| 5 | thal — Thalassemia Type | Nuclear stress test result (normal, fixed defect, reversible defect) | **Reversible defects** indicate areas of the heart that struggle under stress — high-risk pattern |

## The Trust Gap: Why Explainability Matters

A black-box model with 92% ROC-AUC means little to a cardiologist who cannot understand why it flagged a patient.  
Explainability tools bridge this gap:

- **SHAP** provides a rigorous, game-theoretically grounded decomposition of each prediction into feature contributions.
- **LIME** provides a simple, human-readable local approximation: "This patient was predicted high-risk mainly because their ST depression was 2.3 and they had 2 blocked vessels."

Together, they allow clinicians to:
1. Verify that the model relies on clinically plausible features (not data artifacts)
2. Identify unusual cases where the model's reasoning diverges from clinical intuition
3. Build trust with patients who have the right to understand automated decisions (EU AI Act, GDPR Art. 22)

## Limitations and Ethical Considerations

| Limitation | Implication |
|-----------|------------|
| Small dataset (n = 303) | Model may not generalise to diverse populations; needs validation on larger, multi-centre data |
| 1988 Cleveland cohort | Demographics and treatment protocols have changed; feature distributions may differ from modern patients |
| Binary outcome | Heart disease is a spectrum; binary classification oversimplifies clinical reality |
| LIME instability | Different random seeds in LIME can shift feature weights — always cross-check with SHAP |
| No causal inference | SHAP values reflect correlation, not causation — e.g., low thalach could be due to medication, not disease |

## Final Recommendation

> This XGBoost model, combined with SHAP explanations, is suitable for use as a **clinical decision support tool** — not a replacement for physician judgement.  
> For deployment, we recommend: (1) prospective validation on a held-out hospital cohort, (2) SHAP-based explanation reports generated automatically for each patient flagged by the model, and (3) a mandatory clinician review step before any action is taken on the model's output.
